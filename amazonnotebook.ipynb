```python
import string,time
exclude = string.punctuation
import re
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
stopwords.words('english')
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
from nltk.corpus import wordnet
import spacy
```

```python
def remove_html_tags(text):
    return BeautifulSoup(text, 'html.parser').get_text()
```

```python
def replace_punc(text):
    pattern = r'[^\w\s]'
    return re.sub(pattern, ' ', text)
```

```python
def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
nlp = spacy.load("en_core_web_lg")
lemmatizer = WordNetLemmatizer()
text = 'stitching'
def lemmatize(text):
    new_m_list = []
    words = nlp(text)
    lemmatized_words = []
    for word in words:
        # Lemmatize the word 
            lemmatized_words.append(word.lemma_)  # lemmatising word and appending it to the list
    new_sentence = ' '.join(lemmatized_words)
    return new_sentence
print(lemmatize(text))
```

```python
text = 'simple  elegant great  display indoor plant like succulent plant air plant seedle herb etc high quality  indoor outdoor planter stand  flower pot  make  wrought iron material  powder coat paint rustproof  durable pefect decor  live room balcony  garden good quality strong  durable  surface   pot holder  rust proof     use   long time  elevated indoor pot plant stand  make watering easy  give  plant  well space  grow  pragmatic  eco friendly gift   nature lover gardening hobbyist  home decorist note plant    display    include  planter please allow 0 2 cm error due  manual measurement'
def string_with_only_single_spaces(text):
    return re.sub('\s+', ' ',text)
print(string_with_only_single_spaces(text))
```

```python
def stringpreprocessing(chunk, x):
    new_text = ''
    chunk = chunk.fillna('')
    t_my_list = []
    for i in range(chunk.shape[0]):
        if i%500 == 0:
            print(i,'th iteration')
        chunk.iloc[i,x] = replace_punc(remove_html_tags(chunk.iloc[i,x]))
        my_string = re.sub(r'[^\x00-\x7F]+', ' ',chunk.iloc[i,x].lower())
        lemmatized_string = lemmatize(my_string)
        t_my_list.append(string_with_only_single_spaces(remove_stopwords(lemmatized_string)))
    return t_my_list
```

```python
description_total_list = []
bullet_points_total_list = []
i = 0
for chunk in amazon_chunks_train:
    i = i + 1 
    print(i,'th iteration')
    description_list = stringpreprocessing(chunk,2) 
    description_total_list.extend(description_list)
    bullet_points_list = stringpreprocessing(chunk,3)
    bullet_points_total_list.extend(bullet_points_list)
    if i == 30:
        break
```

```python
print(len(bullet_points_total_list))
```

```python
title_total_list = []
i = 0
for chunk in amazon_chunks_train:
    i = i + 1
    title_list = stringpreprocessing(chunk,1) 
    title_total_list.extend(title_list)
    if i == 30:
        break
```

```python
j = 0
new_y = pd.DataFrame()
prod_type_id = pd.DataFrame()
for chunk in amazon_chunks_train:
    j = j + 1
    y = chunk.iloc[:,5]
    product_typeid = chunk.iloc[:,4]
    print(y.shape)
    new_y = pd.concat([new_y, y], ignore_index=True)
    prod_type_id = pd.concat([prod_type_id, product_typeid], ignore_index=True)
    if j == 30:
        break
```

```python
print(prod_type_id.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features =210)
bullet_points = cv.fit_transform(bullet_points_total_list).toarray()
description = cv.fit_transform(description_total_list).toarray()
title = cv.fit_transform(title_total_list).toarray()
```

```python
import pickle
filename = 'cv.pkl'
with open(filename, 'wb') as file:
    pickle.dump(cv, file)
```

```python
desc = pd.DataFrame(description)
bullet_p = pd.DataFrame(bullet_points)
t_title = pd.DataFrame(title)
from sklearn.preprocessing import StandardScaler
scaler3 = StandardScaler()
prod_t_id = scaler3.fit_transform(prod_type_id)
x = pd.concat([desc,bullet_p,t_title,pd.DataFrame(prod_t_id)],axis = 1)
```

```python

from sklearn.model_selection import train_test_split
X_train, X_val, Y_train, Y_val = train_test_split(x, new_y, test_size=0.05, random_state=42)
scaler1 = StandardScaler()
scaler2 = StandardScaler()
x_train = scaler1.fit_transform(X_train)
x_val = scaler1.transform(X_val)


# Fit the PCA model on x_train and transform it
y_train = scaler2.fit_transform(np.array(Y_train).reshape(-1,1))
y_val  = scaler2.transform(np.array(Y_val).reshape(-1,1))
```

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor(n_neighbors=5)
model.fit(x_train,y_train)
y_pred_train = model.predict(x_train)
y_pred_val = model.predict(x_val)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print(mean_absolute_percentage_error(y_val,y_pred_val))
```

```python
score = max( 0 , 100*(1-mean_absolute_percentage_error(y_val,y_pred_val)))
print(score)
```
